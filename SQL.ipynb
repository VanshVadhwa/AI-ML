# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import pandas as pd

print("SURVIVED:",len(df[df["Survived"]==1]))
print("DIED:",len(df[df["Survived"]==0]))


Survived=len(df[df["Survived"]==1])
Died=len(df[df["Survived"]==0])


import matplotlib.pyplot as plt

labels = ['Survived','Died']
sizes = [Survived,Died]
colors = ['lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('SURVIVED AND DIED PEOPLE GRAPH')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

# BIVARIANT ANALSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = df.dropna()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df, color='blue', alpha=0.7)
plt.title('Bivariate Analysis: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df.groupby(['Sex', 'Survived']).size().unstack()
fig, ax = plt.subplots()
ax.pie(survival_counts[1], labels=survival_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'pink'])
ax.set_title('Survival Distribution by Gender')
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df.groupby(['Pclass', 'Survived']).size().unstack()
fig, ax = plt.subplots()
ax.pie(survival_counts[1], labels=survival_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'green','yellow'])
ax.set_title('Survival Distribution by PassengerClass')
plt.show() 
```
